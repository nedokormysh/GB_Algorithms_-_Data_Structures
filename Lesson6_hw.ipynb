Класс для подсчёта объёма памяти. Посчитал, что выводить значения внутренних переменных, формируемых внутри словаря нецелесообразно, но при этом объём занимаемой памяти высчитывается для всех вложенных переменных.

```python
class Memory_Count:
    @staticmethod
    def show_size(x, level=0):
        res_size = sys.getsizeof(x)
        if hasattr(x, '__iter__'):
            if hasattr(x, 'items'):
                for xx in x.items():
                    res_size += Memory_Count.show_size(xx, level+1)
            elif not isinstance(x, str):
                for xx in x:
                    res_size += Memory_Count.show_size(xx, level+1)
        return res_size

    @staticmethod
    def memory_count(locals):
        memory_on_var = 0
        print('\nВ данной программе используются переменные: ')

        i = 1

        for el in locals.keys():
            memory_value = Memory_Count.show_size(locals[el])
            if hasattr(el, '__iter__') and not isinstance(locals[el], int):
                print(f'{i}) переменная {el} имеет тип: {type(locals[el])} и значение: {locals[el]}. Вместе с вложенными в неё переменными занимает объём памяти: {memory_value} байт(-a)')
            else:
                print(f'{i}) переменная {el} имеет тип: {type(locals[el])} и значение: {locals[el]}. Занимает объём памяти: {memory_value} байт(-a)')
            memory_on_var += Memory_Count.show_size(locals[el])
            i += 1

        print(f'\033[1mОбщий объём памяти, выделенный под программу: {memory_on_var} байт(-а)\033[0m')
```

## Вариант 1.

```python
def count_v1():
    for i in range(2, 10):
        count = 0
        for j in range(2, 100):
            if j % i == 0:
                count += 1
        print(f'Количество чисел делящихся на {i} = {count}')

    Memory_Count.memory_count(locals())


count_v1()
```

Вариант 1 занимает относительно мало места за счёт объявления только трёх целочисленных переменных. 

## Вариант 2.

```python
def count_v2():
    count = []
    for i in range(2, 10):
        c = 0
        for j in range(2, 100):
            if j % i == 0:
                c += 1
        count.append(c)

    for k in range(2, len(count) + 2):
        print(f'Количество чисел делящихся на {k} = {count[k - 2]}')

    Memory_Count.memory_count(locals())


count_v2()
```

Во втором варианте создаём 5 переменных. В теории могли уменьшить количество переменных за счёт использования во втором цикле ранее объявленной, например, j вместо k.

## Вариант 3.

```python
def count_v3():

    count = {}

    for i in range(2, 10):
        count[i] = 99 // i
        print(f'Количество чисел делящихся на {i} = {count[i]}')

    Memory_Count.memory_count(locals())


count_v3()
```
