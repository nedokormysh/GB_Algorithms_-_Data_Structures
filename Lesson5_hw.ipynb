# Задание 1.

Пользователь вводит данные о количестве предприятий, их наименования и прибыль за четыре квартала для каждого предприятия. Программа должна определить среднюю прибыль (за год для всех предприятий) и отдельно вывести наименования предприятий, чья прибыль выше среднего и ниже среднего.

```python
from collections import namedtuple

while True:
    try:
        n = int(input('Введите количество компаний: '))
    except ValueError:
        print('Вы ввели недопустимое значение. Введите одно натуральное число.')
        continue
    break

lst_set = ['name', 'year_income']
Company = namedtuple('Company', lst_set)
companies_dct = {}
companies = []
q = 4
i = 0

while i < n:
    while True:
        try:
            s = input('Введите название фирмы и поквартальную прибыль через пробел: ').split()

            if len(s) != 1 + q:
                print('Ввели недопустимое количество элементов. Требуется ввести название фирмы и 4 цифры через пробел.')
                continue

            year_income = 0

            for j in range(1, len(s)):
                j = float(s[j])
                year_income += j

            s[0] = Company(s[0], year_income)
            companies_dct[s[0].name] = s[0].year_income

        except ValueError:
            print('Вы ввели недопустимое значение. Проверьте значения поквартальных прибылей. Эти поля должны быть числовыми.')
            continue
        break

    i += 1

sum_year_income = 0

for name, year_income in companies_dct.items():
    print(f'Предприятие {name} имеет прибыль {year_income}')
    sum_year_income += year_income

middle_year_income = sum_year_income / n

print(f'\033[1mСредняя годовая прибыль компаний: {round(middle_year_income, 2)} \033[0m')

print('\033[1mПредприятия, у которых прибыль за год больше, чем в среднем: \033[0m')

for name, year_income in companies_dct.items():
    if year_income > middle_year_income:
        print(f'{name} - {year_income}')


print('\033[1mПредприятия, у которых прибыль за год меньше, чем в среднем: \033[0m')

for name, year_income in companies_dct.items():
    if year_income < middle_year_income:
        print(f'{name} - {year_income}')
```

# Задание 2.

Написать программу сложения и умножения двух шестнадцатеричных чисел. При этом каждое число представляется как массив, элементы которого — цифры числа.
Например, пользователь ввёл A2 и C4F. Нужно сохранить их как [‘A’, ‘2’] и [‘C’, ‘4’, ‘F’] соответственно. Сумма чисел из примера: [‘C’, ‘F’, ‘1’], произведение - [‘7’, ‘C’, ‘9’, ‘F’, ‘E’].
Примечание: Если воспользоваться функциями hex() и/или int() для преобразования систем счисления, задача решается в несколько строк. Для прокачки алгоритмического мышления такой вариант не подходит. Поэтому использование встроенных функций для перевода из одной системы счисления в другую в данной задаче под запретом.
Вспомните начальную школу и попробуйте написать сложение и умножение в столбик.

```python
from collections import deque
from copy import copy

hex_dig = '0123456789ABCDEF'

class HexCalcSimple:

    hex_dig = '0123456789ABCDEF'

    def __init__(self, num):
        self.num = deque(num)

    @staticmethod
    def sum_digits(a, b, k):

        j = 0
        s = hex_dig.index(a) + hex_dig.index(b) + k

        if s >= 16:
            j = 1

        return hex_dig[s % 16], j

    @staticmethod
    def mul_digits(a, b, k):

        j = 0
        s = hex_dig.index(a) * hex_dig.index(b) + k

        if s > 15:
            j = s // 16

        return hex_dig[s % 16], j

    def __add__(self, other):

        if len(self.num) >= len(other.num):
            a_ = copy(self.num)
            b_ = copy(other.num)
        else:
            a_ = copy(other.num)
            b_ = copy(self.num)

        b_.extendleft('0' * (len(a_) - len(b_)))

        result = deque()

        k = 0

        for i in range(len(a_) - 1, -1, -1):
            res_, k = HexCalcSimple.sum_digits(a_[i], b_[i], k)
            result.appendleft(res_)

        if k != 0:
            result.appendleft(k)
        result = HexCalcSimple(result)

        return result

    def __str__(self):
        s = ''
        for i in self.num:
            s += str(i)
        return s

    def __mul__(self, other):
        a_ = copy(self.num)
        b_ = copy(other.num)

        result_temp = deque()

        d = 0
        result_mult = HexCalcSimple('0')
        res_t = ''

        for i in range(len(a_) - 1, -1, -1):
            k = 0
            for j in range(len(b_) - 1, -1, -1):
                res_, k = HexCalcSimple.mul_digits(a_[i], b_[j], k)
                result_temp.appendleft(res_)

            if k != 0:
                result_temp.appendleft(hex_dig[k])

            for m in range(d):
                result_temp.append('0')

            for t in result_temp:
                res_t += str(t)

            result_temp.clear()

            res_t = HexCalcSimple(res_t)
            result_mult = result_mult + res_t
            res_t = ''
            d += 1

        return result_mult

def Main():

    hex_dig = '0123456789ABCDEF'

    i = 0
    numbers = []

    while i < 2:
        num = input(f'Введите {i + 1}-ое 16-ричное число: ').upper()

        control = 1

        for digit in num:
            if digit not in hex_dig:
                control = 0

        if control:
            i += 1
            numbers.append(num)

        else:
            print(f'Введено не 16-ричное число. Ошибка числе {i + 1}')
            continue

    x = numbers[0]
    y = numbers[1]

    print(f'Вы ввели x = {x} и y = {y}')

    x = HexCalcSimple(x)
    y = HexCalcSimple(y)

    print(f'x + y = {x + y}')
    print(f'x * y = {x * y}')

Main()
```
